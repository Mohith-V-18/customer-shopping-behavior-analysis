```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
df.head(10)
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)':'purchse_amount'})
df.columns
```

```python
bins = [0, 18, 30, 55, 100]
labels = ['Young', 'Adults', 'Middle-Aged', 'Seniors']

df['age_groups'] = pd.cut(df['age'], bins=bins, labels=labels)
df[['age', 'age_groups']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Every 3 Months' : 90}

df['purchase_frequency'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency']].head(10)
```

```python
df.isnull().sum()
```
